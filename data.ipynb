```python
import pandas as pd
import numpy as np
import json
```

```python
data = pd.read_csv('pins.csv')
display(data)
print(data.shape)

```

```python
newPins = []

for index, row in data.iterrows():
    entry = {
        'id': index + 1,
        'artist': row['artistName'],
        'song': row['trackTitle'],
        'location': row['neighborhood'],
        'lat': row['latitude'],
        'lng': row['longitude'],
        'audio_url': row['trackStreamUrl']
        }
    newPins.append(entry)
```

```python
with open('pins.json', 'w') as f:
    json.dump(newPins, f, indent=4)
    
```
